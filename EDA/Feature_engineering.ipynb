# Feature Engineering

- Age Category
- BMI Category - Underweight, normal Weight, Overweight, Obesity
- Pollution Risk - Location and Air Pollution level
- Smoking Status Encoding
- Interaction Features
- Location Encoding

```python
#importing necessary libraries
import numpy as np
import pandas as pd
```

```python
#Loading the dataset
df = pd.read_csv("G:\DataScience\AIcourse\COPD Prediction\Data\synthetic_COPD_data.csv")
df.head()
```

```python
#Creating Age categories
df['Age_Category']= pd.cut(df['Age'], bins=[29, 39, 49, 59, 69, 79,89], labels = ['30 - 39', '40 - 49', '50 - 59', '60 - 69', '70 - 79', '80-89'])

df.head()
```

```python
#Creating BMI Categories
df['BMI_Categories'] = pd.cut(df['BMI'], bins = [0, 18.5, 24.5, 29.9, 35], labels= ['Underweight', 'Normal Weight', 'Overweight', 'Obesity'])

df.head()
```

```python
# Encoding Pollution Risk Score
df['Pollution_Risk_Score'] = np.where(df['Air_Pollution_Level'] > 150, 1, 0)
```

```python
# Encoding smoking status
df['Smoking_Status_encoded'] = df['Smoking_Status'].map({'Current': 1, 'Former': 0.5, 'Never': 0})
```

```python
# encoding gender
df['Gender_'] = df['Gender'].map({'Male': 1, 'Female':0})
```

```python
# Interaction Features - Smoking status and pollution level
df['Smoking_Pollution_interaction'] = df['Smoking_Status_encoded'] * df['Air_Pollution_Level']
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Location'], drop_first=True)
```
