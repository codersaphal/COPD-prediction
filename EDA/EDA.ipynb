# COPD Prediction EDA

## Factors that lead to COPD
1. Biomass Fuel Exposure (Indoor AIr pollution)
2. Tobacco Smoking
3. Outdoor Air pollution
4. Occupational Exposure
5. Genetic susceptability
6. Respiratory Infections
7. Low Awareness and health Access
8. Climate and Environmental factors

## Signs and symptoms of COPD
1. Chronic Cough
2. Shortness of Breath (Dyspnea)
3. Excess mucus production (Sputum) (Phlegm)
4. Wheezing
5. Fatigue and low energy
6. Chest tightness
7. Frequent Respiratory Infections
8. Cynosis (Bluish lips or Fingernail beds)
9. Unintended Weight
10. Chronic Bronchitis and Emphysema Symptoms
11. Frequent Flare-ups (Exacerbations)

# Data Visualization

```python
#Setting to large plot ensuring better visibility
plt.figure(figsize=(16,9))
```

# UniVariate Analysis

```python
#Visualizing COPD_Diagnosis
sns.countplot(data=df,x='COPD_Diagnosis')
plt.title("COPD_Diagnosis Distribution")
plt.show()
```

```python
#visualizing age distribution
sns.histplot(x='Age',data=df, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Visualizing BMI distribution
sns.histplot(data=df,x='BMI', kde=True)
plt.title("BMI Distribution")
plt.show()
```

```python
#gender distribution
sns.countplot(data=df, x='Gender')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(data=df,x="Smoking_Status")
plt.title("Smoking Status Distribution")
plt.show()
```

# BiVariate Analysis

```python
#Smoking vs COPD Diagnosis
sns.countplot(data=df, x='Smoking_Status', hue='COPD_Diagnosis')
plt.title("Smoking_status vs COPD_Diagnosis")
plt.show()
```

Thus, we can conclude that smoking is one of the major causes of COPD. 

```python
#LocationvsAir_PollutionLEvel
sns.barplot(x='Location',y='Air_Pollution_Level',hue="COPD_Diagnosis",data=df, ci=False)
plt.title("Air Pollution Levels Across Cities by COPD Prediction")
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python
#agevscopddiagnosis
sns.boxplot(y='Age',x='COPD_Diagnosis',data=df)
plt.title("Age vs COPD_Diagnosis")
plt.show()
```

```python
#BMIvsdiagnosis
sns.boxplot(y='BMI',x='COPD_Diagnosis',data=df)
plt.title("BMI vs COPD Diagnosis")
plt.show()
```

```python
sns.countplot(hue='COPD_Diagnosis',x='Family_History_COPD',data=df)
plt.title("Family History vs COPD Diagnosis")
plt.show()
```

Hence, Family_history on COPD can also be a major cause in COPD 

```python
sns.violinplot(data=df, x='COPD_Diagnosis', y='Air_Pollution_Level',kde=True)
plt.title('Air Pollution Level vs COPD Diagnosis')
plt.show()
```

```python
#GendervsCOPD_Diagnosis
sns.countplot(data=df,x='Gender',hue="COPD_Diagnosis")
plt.title("Gender vs COPD_Diagnosis")
plt.show()
```

Gender donot have any affect on COPD. It can occur in any of them

```python
#Occupational_exposurevsCOPD_Diagnosis
sns.countplot(data=df,x='Occupational_Exposure',hue="COPD_Diagnosis")
plt.title("Occupational Exposure vs COPD_Diagnosis")
plt.show()
```
